```python
import pandas as pd

data_pokemon={
    'Nama':['Charizard','Squirtle','Bulbasaur','Pikachu','Onix','Geodude','Vulpix','Starmie'],
    'Tipe':['Fire','Water','Grass','Electric','Rock','Rock','Fire','Water'],
    'HP':[78,44,45,35,35,40,38,60],
    'Attack':[84,48,49,55,45,80,41,75],
    'Defense':[78,65,49,40,160,100,40,85]
}
df=pd.DataFrame(data_pokemon)
df.to_csv('pokemon.csv',index=False)
df
```

```python
rata_serangan=df.groupby('Tipe')['Attack'].mean()
print(rata_serangan)
df.groupby('Tipe')['Defense'].count()
df.groupby('Tipe')['Attack'].mean().sort_values(ascending=False)
```
