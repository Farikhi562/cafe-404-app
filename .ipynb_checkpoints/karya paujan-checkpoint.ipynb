```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize']=(10,6)

data_real={
    'Mata_kuliah':['Algoritma pemrogaman','Matematika dasar','Bahasa Inggris','Matematika Informatika','Pengenalan Teknologi komputer',
                   'Bahasa Indonesia','PKN','Fisika dan Kimia dasar'],
    'SKS':[3,2,2,2,2,2,2,2],
    'Tugas_Avg':[86,90,100,90,80,91,90,93],
    'Quiz_Avg':[81,95,65,97,75,82,83,91],
    'Absensi':[90,90,100,100,83,80,68,85],
    'UTS_Murni':[87,80,70,100,81,85,100,85],
    'Remedial':[0,0,84,0,0,0,0,0],
}
df=pd.DataFrame(data_real)
print('Dataset berhasildimuat (User data).')
display(df)
```

```python
bobot_tugas=0.30
bobot_quiz=0.50
bobot_absen=0.20
bobot_akhir_proses=0.20
bobot_akhir_uts=0.45
bobot_akhir_uas=0.35
print('Konfigurasi bobot dimuat..')

cols_numeric=['Tugas_Avg','Quiz_Avg','Absensi','UTS_Murni','Remedial']
df[cols_numeric]=df[cols_numeric].clip(lower=0,upper=100)

df['UTS_Final']=np.where(df['Remedial']>df['UTS_Murni'],
                         df['Remedial'],
                        df['UTS_Murni']
                        )
df['Status_Remed']=np.where(df['Remedial']>0,'Ya','Tidak')
df['Nilai_Proses']=(
    (df['Tugas_Avg']*bobot_tugas)+(df['Quiz_Avg']*bobot_quiz)+(df['Absensi']*bobot_absen)
)
#Rumus: (nilai proses x 20%)+(UTS Finalx45)
df['Modal_sebelum_UAS']=(
    (df['Nilai_Proses']*bobot_akhir_proses)+
    (df['UTS_Final']*bobot_akhir_uts)
)

#Risk assesment (DEteksi bahaya)
#menambahkan kategori otomatis buat visualisasi
#Asumsi: jika modal<40, bahaya karena beban uas jadi berat bgt
def cek_status(modal):
    if modal>=55: return 'Sangat Aman' 
    elif modal>=45: return 'Aman'
    elif modal>=35: return 'Warning!'
    else: return 'Bahaya (zona merah)'
df['Status_Matkul']=df['Modal_sebelum_UAS'].apply(cek_status)
print('Data cleaning & Feature Engineering Selesai.')
print('   - Outlier values dikoreksi (clipping 0-100).')
print('   - Logika Remedial diaplikasikan.')
print('   - Risk Assesment ditambahkan.')
cols_display=['Mata_kuliah','UTS_Murni','Remedial','UTS_Final','Nilai_Proses','Modal_sebelum_UAS','Status_Matkul']
display(df[cols_display].sort_values(by='Modal_sebelum_UAS',ascending=True))
```

```python
#Data real time kita cuma 6 baris, model ML butuh ribuan data biar pinter
#buat 2000 data mahasiswa fiktif dengan pola yg realistis
from sklearn.utils import shuffle

np.random.seed(42)
n_samples=5000
tugas_syn=np.random.normal(loc=78,scale=12,size=n_samples).clip(0,100)
quiz_syn=np.random.normal(loc=72,scale=15,size=n_samples).clip(0,100)
absen_syn=np.random.choice(
    [100,90,85,80,50,0],
    size=n_samples,
    p=[0.7,0.15,0.05,0.05,0.03,0.02]
)
#Realistisnya nilai  ujian berkorelasi sama tugas 
#tapi ada faktor kaya sakit pas ujian,dosen killer,atau lg hoki
noise_factor=np.random.normal(0,8,n_samples)
uts_syn=(0.6*tugas_syn)+(0.4*np.random.normal(70,15,n_samples))+noise_factor
uas_syn=(0.6*quiz_syn)+(0.4*np.random.normal(65,15,n_samples))+noise_factor

uts_syn=uts_syn.clip(0,100)
uas_syn=uas_syn.clip(0,100)
df_train=pd.DataFrame({
    'Tugas':tugas_syn,
    'Quiz':quiz_syn,
    'Absen':absen_syn,
    'UTS':uts_syn,
    'UAS':uas_syn
})

#buat kunci jawaban AI
#Rumus: 10% Tugas + 15% Quiz + 5% Absen + 40% UTS + 30% UAS
df_train['Nilai_Akhir_Asli']=(
    (df_train['Tugas']*0.10)+
    (df_train['Quiz']*0.15)+
    (df_train['Absen']*0.05)+
    (df_train['UTS']*0.40)+
    (df_train['UAS']*0.30)
)
df_train['Nilai_Akhir_Asli']+=np.random.normal(0,1,n_samples)
df_train['NIlai_Akhir_Asli']=df_train['Nilai_Akhir_Asli'].clip(0,100)

#Cek datanya buatan kita terlihat wajar 
print(f'Berhasil membuat {n_samples} data latih sintetis.')
plt.figure(figsize=(12,5))

#1. plot 1 distribusi nilai akhir
plt.subplot(1,2,1)
sns.histplot(df_train['Nilai_Akhir_Asli'],kde=True,color='purple',bins=30)
plt.title('Distribusi Nilai Akhir Mahasiswa')
plt.xlabel('Nilai Akhir')
plt.ylabel('Jumlah Mahasiswa')

#plot 2 Hubungan tugas vs UAS (cek korelasi)
plt.subplot(1,2,2)
sns.scatterplot(x='Tugas',y='UTS',data=df_train,alpha=0.3,color='teal',s=10)
plt.title('Sebaran Data: Tugas vs UTS')
plt.xlabel('Nilai Tugas')
plt.ylabel('Nilai UTS')
plt.tight_layout()
plt.show()
display(df_train.head())
```

```python
#Ngelatih otak AI, suruhcari pola hubungan antara tugas,quiz,absen,ujian--->Nilai Akhir
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Membagi data
#definisiin fiitur x dan target y
features=['Tugas','Quiz','Absen','UTS','UAS']
target='Nilai_Akhir_Asli'
x=df_train[features]
y=df_train[target]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(f'   Data Split Berhasil:')
print(f'   Data Latih (Training):  {x_train.shape[0]} mahasiswa')
print(f'   Data Uji (Testing):     {x_test.shape[0]} mahasiswa')

model=LinearRegression()
#proses training
print('\n Sedang melatih model...',end="")
model.fit(x_train,y_train)
print('Selesai!')

#Apakah Ai berhasil menebak rumus rahasia yg dibuat di cell 4?
#Rumus Asli cell 4: Tugas(0.10), Quiz (0.15), Absen (0.05), UTS (0.40), UAS (0.30)
coef_data=pd.DataFrame({
    'Komponen':features,
    'Bobot_Asli_Dosen':[0.10,0.15,0.05,0.40,0.30],
    'Tebakan_AI':model.coef_
})
coef_data['Selisih']=abs(coef_data['Bobot_Asli_Dosen']-coef_data['Tebakan_AI'])
print('\n  LAPORAN KECERDASAN MODEL....')
display(coef_data.style.background_gradient(cmap='Greens',subset=['Tebakan_AI']))
print('\nKesimpulan:')
if coef_data['Selisih'].mean()<0.05:
    print('  LUAR BIASA! AI berhasil memecahkan rumus penilaian dosen dengan sangat akurat.')
    print('Perhatikan kolom Tebakan_AI hampir sama persis dengan Bobot_Asli')
else:
    print('Model masih bingung. Mungkin data latih kureng banyak atau terlalu banyak noise.')
```

```python
#Validasi Akurasi
#Tahap ini tujuannya 'Seberapa jago sih AI nebak nilai?'
#bandingin kunci jawaban (y_test) sama tebakan AI (y_pred)
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
#1. Prediksi Data Testing
#Ai disuruh ngerjain soal ujian (Data Test 20%)
y_pred=model.predict(x_test)

#Hitung skor akurasi
score_r2=r2_score(y_test,y_pred)
#Mean absolute error (MAE) adalah rata2 kesalahn tebakan (dlm poin)
score_mae=mean_absolute_error(y_test,y_pred)
#Mean squared error (MSE): Kesalahan kuadrat 
score_mse=mean_squared_error(y_test,y_pred)
print('LAPORAN PERFORMA MODEL.')
print('-'*40)
print(f'R-Squared ($R^2$): {score_r2:.4f}')
print(f' Artinya: Rata-rata tebakan AI meleset sekitar {score_mae:.2f} poin.')

#Visualisasi performa
plt.figure(figsize=(14,6))
#grafik 1: prediction vs aktual
#kalo titik-titik nempel di garis merah berarti prediksi ny sempurna
plt.subplot(1,2,1)
sns.scatterplot(x=y_test,y=y_pred,alpha=0.5,color='blue',edgecolor='k')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--',lw=3,label='Perfect Prediction')
plt.title('Realita vs Prediksi ()Harusnya garis lurus).',fontsize=12,fontweight='bold')
plt.xlabel('Nilai Asli (y_test)')
plt.ylabel('Prediksi AI (y_pred)')
plt.legend()

#Grafik 2. Redisual Analysis
#Kita gcek selisih (error), harusnya si error nyebar nol di angka 0
residuals=y_test-y_pred
plt.subplot(1,2,2)
sns.histplot(residuals,kde=True,color='crimson',bins=30)
plt.axvline(x=0,color='black',linestyle='--')
plt.title('Distribuse Error',fontsize=12,fontweight='bold')
plt.xlabel('Besar Kesalahan (Residuals)')
plt.tight_layout()
plt.show()

#Kesimpulan Akhir
print('\nKESIMPULAN VALIDASI')
if score_r2>0.95:
    print('GRADE A: Model sangat akurat! siap digunakan untuk prediksi IPK.')
elif score_r2>0.80:
    print('GRADE B: Model cukup bagus, tapi masih ada sediki error.')
else:
    print('GRADE C: Model kurang akurat. Perlu tambah data atau ganti algoritma.')
```

```python
#gw akan menerapkan model AI ke data asli
#Fitur spesial: ngitung target minimal UAS utk dapet nilai A
#Samain nama kolom data asli sama data latih AI
input_user=df[['Tugas_Avg','Quiz_Avg','Absensi','UTS_Final']].copy()
input_user.columns=['Tugas','Quiz','Absen','UTS']

#kalo gw kerjain uas dan dapet nilai 75, nilai akhir gw bakal brp?
target_uas=75 #ganti2 aja ini bebas
input_user['UAS']=target_uas
df['Prediksi_Skor_Akhir']=model.predict(input_user)
df['Skenario_UAS_Dipilih']=target_uas

#Skenario 2 Reverse engineering
#berapa nilai UAS minimal buat dapet grade A (85) atau B (70)
#regresi linear:y=b0+b1*x1+...+b_uas*x_uas
#maka x_uas=(Target-(b0+b1*x1...))/b_uas
intercept=model.intercept_
coefs=dict(zip(features,model.coef_))
bobot_uas_ai=coefs['UAS']

#Hitung modal non-UAS kaya dari tugas,quiz,absen,uts
#Rumus=y_partial=b0+(b_tugas*Tugas)+....+(b_uts*UTS)
df['Modal_Regresi']=(
    intercept+
    (input_user['Tugas']*coefs['Tugas'])+
    (input_user['Quiz']*coefs['Quiz'])+
    (input_user['Absen']*coefs['Absen'])+
    (input_user['UTS']*coefs['UTS'])
)
def hitung_target_uas(target_nilai_akhir,modal,bobot_uas):
    #Rumus aljabar: UAS+(Nilai_Akhir-Modal)/Bobot_UAS
    butuh=(target_nilai_akhir-modal)/bobot_uas
    return butuh
#Terapin buat nyari Target A (misal A=85)
df['butuh_uas_A']=df['Modal_Regresi'].apply(lambda x: hitung_target_uas(85,x,bobot_uas_ai))

#Terapin buat nyari B (misal B=  >70)
df['butuh_uas_B']=df['Modal_Regresi'].apply(lambda x: hitung_target_uas(70,x,bobot_uas_ai))

#Terapin buat cari C (misal C=  >60)
df['butuh_uas_C']=df['Modal_Regresi'].apply(lambda x: hitung_target_uas(60,x,bobot_uas_ai))

print(f'SIMULASI SELESAI (Skenario UAS lu : {target_uas} {nama}!)')
print('-'*60)

cols_show=['Mata_kuliah','Modal_sebelum_UAS','Prediksi_Skor_Akhir','butuh_uas_A','butuh_uas_B','butuh_uas_C']
pd.options.display.float_format='{:.1f}'.format
#Tambahin logic status apa mungkin gw dapet A?
#kalo butuh UAS>100, berartii mustahil dapet A (kecuali dosen kasi bonus)
def cek_kemungkinan(nilai_butuh):
    if nilai_butuh>100: return "Mustahil dapet A (Max>100)"
    elif nilai_butuh<0: return "Auto A (Tidur pun A)"
    else: return f"Butuh {nilai_butuh:.0f}"
df['Status_Target_A']=df['butuh_uas_A'].apply(cek_kemungkinan)
display(df[['Mata_kuliah','Prediksi_Skor_Akhir','butuh_uas_A','Status_Target_A']])
```

```python
#Final report & GPA Calculator
#konversi angka ke huruf dan hitung IP semester.

#definisi aturan grade (sesuai kampus)
def konversi_nilai(skor):
    if skor>=85: return 'A', 4.0  #Sangat memuaskan
    elif skor>=70: return 'B', 3.5 #Memuaskan
    elif skor>=60: return 'C', 3.0 #Cukup
    elif skor>=45: return 'D', 2.7 #Kureng
    else: return 'E', 0.0  #Gagal

hasil_konversi=df['Prediksi_Skor_Akhir'].apply(lambda x:pd.Series(konversi_nilai(x),index=['Grade_Prediksi','Bobot_SKS']))
final_report=pd.concat([df,hasil_konversi],axis=1)

#Ngitung IP semester
#Rumus IP=(total bobotxSKS)/Total sks
#itung mutu per matkul (SKS*bobot)
final_report['Total_Mutu']=final_report['SKS']*final_report['Bobot_SKS']
total_sks_diambil=final_report['SKS'].sum()
total_poin_mutu=final_report['Total_Mutu'].sum()
ip_semester=total_poin_mutu/total_sks_diambil

print('=================================================================')
print(f'   PREDIKSI HASIL STUDI SEMESTER 1 (Skenario UAS: {target_uas})')
print('=================================================================')
cols_final=['Mata_kuliah','SKS','Modal_sebelum_UAS','Prediksi_Skor_Akhir','Grade_Prediksi','Bobot_SKS']
display(final_report[cols_final])
print('\n'+'-'*50)
print(f'Nama Mahasiswa: {nama}')
print(f'TOTAL SKS DIAMBIL: {total_sks_diambil}')
print(f'PREDIKSI IP SEMESTER: {ip_semester:.2f}')
print('-'*50)

#AI ADVISOR SARAN OTOMATIS
print('\n AI ADVISOR COMMENTS:')
if ip_semester>=3.50:
    print(' PERFORMANCE: CUMLAUDE LEVEL!')
    print(f' Saran: Pertahankan! Lu berpotensi jadi asisten lab atau dapet beasiswa {nama}.')
elif ip_semester>=3.25:
    print(' PERFORMANCE: GOOD JOB.')
    print(f' Saran: Aman. Tapi hati-hati di matkul yang Grade-nya masih B/C, coba push lagi {nama}.')
elif ip_semester>=2.70:
    print(' PERFORMANCE: WARNING.')
    print(f' Saran: IP segini mah rawan. Fokus perbaikin matkul dengan SKS besar (3/4) SKS agar ip nya ke dongkrak {nama}.')
else:
    print(' PERFORMANCE: DANGER ZONE.')
    print(f' Saran: Segera konsultasi ke Dosen Wali. Pertimbangkan Semester Pendek (SP) untuk mengulang {nama}.')
    
```

```python
#tahap ini akan menjawab: matkul apa yg harus gua pelajarin duluan?
#dan nyimpen semua hasil kerja ke file
import joblib

#Skala prioritas
#Logika AI: Prioritas=(SKS gede) AND (Modal masih kecil)
df['Gap_to_A']=85-df['Modal_sebelum_UAS']
df['Gap_to_A']=df['Gap_to_A'].apply(lambda x: x if x>0 else 0) #kalo udh A gapnya 0
#semakin gede skornya semakin wajib lu belajar matkul ini
#Skor prioritas=Gap*SKS
df['Skor_Prioritas']=df['Gap_to_A']*df['SKS']

#urutin dri yg paling penting
prioritas_belajar=df.sort_values(by='Skor_Prioritas',ascending=False)

print('DAFTAR PRIORITAS BELAJAR (URUTAN WAJIB)!')
print('(AI menyarankan lu belajar berurutan dari atas ke bawah.)')
print('-'*60)
cols_strategy=['Mata_kuliah','SKS','Modal_sebelum_UAS','Gap_to_A','Status_Target_A']
display(prioritas_belajar[cols_strategy])

#Export laporannya ke excel/csv
nama_file_laporan='Laporan_Prediksi_IPK.csv'
final_report.to_csv(nama_file_laporan,index=False)
print(f'\n Laporan berhasil disimpan ke file: {nama_file_laporan} ya {nama}')
print(f'    Cek folder project lu coba {nama}, file ini bisa di buka di excel.')

#Save model AI (ML OPS)
nama_file_model='otak_ai_prediksi_nilai.pkl'
joblib.dump(model,nama_file_model)
print(f' Model AI berhasil disimpen ke {nama_file_model} yakkk {nama}')
print(f' Lu sekarang punya file otak buatan lu sendiri {nama}!')
```
